```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os


# assign directory
pasta = "Grficos"
i = 0

for arquivo in os.listdir(pasta):
    f = os.path.join(pasta, arquivo)
    # checking if it is a file
    if os.path.isfile(f):
    
        #for file in 
        data = np.loadtxt(f, skiprows = 55)
        i = i + 1

        dataX = []
        dataY = []
        for linha in data:
            dataX.append(linha[0])
            dataY.append(linha[1])
        
        plt.figure(i)
        plt.plot(dataX,dataY)

        plt.savefig(os.path.join(pasta,arquivo[:8]) + '.png')
        #print(arquivo[:8] + '.txt')

```
