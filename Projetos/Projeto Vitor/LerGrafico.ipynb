```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
plt.rcParams.update({
    #"axes.unicode_minus" : False,
    "font.family": "serif",  # use serif/main font for text elements
    "font.serif" : "Computer Modern Roman",
    "font.sans-serif" : "Computer Modern Sans serif",
    "font.monospace"  : "Computer Modern Typewriter",
    "font.weight"   :   "bold",
    #"text.usetex": True,     # use inline math for ticks
    #"pgf.rcfonts": False,    # don't setup fonts from rc parameters   
    "figure.figsize": [10,5],
    #"axes.titlesize": "xx-large",   # fontsize oarqthe axes title
    #"axes.labelsize": "xx-large",  # fontsize of the x any y labels
    #"xtick.labelsize": "xx-large",  # font size of the tick labels
    #"ytick.labelsize": "xx-large",  # font size of the tick labels
    #"legend.fontsize":   "large",
    #"pgf.preamble": "\n".join([
         #r"\usepackage{url}",            # load additional packages
         #r"\usepackage{unicode-math}",   # unicode math setup
         #r"\setmainfont{DejaVu Serif}",  # serif font via preamble

    #])
})


# assign directory
pasta = "Gráficos"
i = 0

for arquivo in os.listdir(pasta):
    if arquivo[-3:] == "asc":
        f = os.path.join(pasta, arquivo)
        # checking if it is a file
        if os.path.isfile(f):
            titulo = open(f, "r").readlines()[8][:-24]

            nome = np.loadtxt(f, skiprows = 55)
            #for file in 
            data = np.loadtxt(f, skiprows = 55)
            i = i + 1

            dataX = []
            dataY = []
            for linha in data:
                dataX.append(linha[0])
                dataY.append(linha[1])

            plt.figure(i)
            plt.plot(dataX,dataY)
            plt.xlabel(r'Número de onda ($cm^{-1})$', fontweight='bold')
            plt.ylabel('Reflectância (u.a)', fontweight='bold')
            plt.xlim(dataX[0],dataX[-1])
            plt.ylim(min(dataY),max(dataY)+1)
            plt.title(titulo + ' - ' + arquivo[0:-4])

            #plt.savefig(os.path.join(pasta,arquivo[:8]) + '.jpg')
            #plt.savefig(os.path.join(pasta,arquivo[:8]) + '.png')
            #print(arquivo[:8] + '.txt')
            plt.close()


def PlotarGrafico(arq,titulo):
    pasta = "Gráficos"

    f = os.path.join(pasta, arq  + ".asc")
    # checking if it is a file
    if os.path.isfile(f):

        #for file in 
        data = np.loadtxt(f, skiprows = 55)

        dataX = []
        dataY = []
        for linha in data:
            dataX.append(linha[0])
            dataY.append(linha[1])

        fig,ax = plt.subplots(figsize = [16,8])
        ax.plot(dataX,dataY)
        ax.set_xlim(dataX[0],dataX[-1])
        ax.set_xlabel(r'Número de onda ($cm^{-1})$', fontweight='bold')
        ax.set_ylabel('Reflectância (u.a)', fontweight='bold')
        ax.set_ylim(min(dataY),max(dataY)+1)
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)

        if titulo:
            plt.title(arq)

        return [fig,ax]

def SalvarGrafico(fig,arq,close):
        fig.savefig(os.path.join(pasta, arq) + '.jpg')
        fig.savefig(os.path.join(pasta, arq) + '.png')
        if not close:
            plt.close(fig)
```
